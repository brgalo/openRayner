```python
from random import seed, random
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.arccos(np.random.random(100))
print(np.mean(arr))
cos = np.cos(arr)
sin = np.sin(arr)

plt.plot(sin,cos,".")
plt.xlim(0,1.1)
plt.ylim(0,1.1)
ax = plt.gca()
ax.set_aspect('equal')

```
